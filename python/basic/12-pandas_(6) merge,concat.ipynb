```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
df1 = DataFrame(np.arange(1,7).reshape(2,3), columns=['A','B','C'])
df1
```

```python
df2 = DataFrame(np.arange(10,61,10).reshape(2,3), columns=['A','B','C'])
df2
```

# concat

```python
pd.concat([df1, df2]) #default: axis=0, 행의 결합 (세로 방향으로 합쳐짐)
```

```python
pd.concat([df1,df2], axis = 0)
```

```python
pd.concat([df1,df2], axis = 1)  #열의 결합
```

```python
pd.concat([df1, df2], ignore_index=True) #순차적인 인덱스 부여
```

# 조인
- 두 데이터프레임(테이블) 참조조건을 활용하여 하나의 객체로 합치거나 데이터를 처리하는 행위
- merge가 두 데이터프레임 조인을 수행
- 등가조건만을 사용하여 조인 가능

##### pd.merge(left,right,how='inner',on=,left_on=,right_on=)

- inner join

```python
pd.merge(emp, df_dfpt, on='deptno')  
```

- outer join

```python
df_dept_new = DataFrame({'deptno':[10,20],'dname':['인사총무부','IT분석팀']})
df_dept_new
```

```python
pd.merge(emp, df_dept_new, on='deptno')  #inner : 30번 부서원 생략
```

```python
pd.merge(emp, df_dept_new, on='deptno', how = 'outer') #outet : NaN으로 함께 출력
```

```python
pd.merge(emp, df_dept_new, on='deptno', how = 'left')
```
