### sort() 정렬
1. sort_index
- Series, DataFrame 호출 가능
- index, column 재배치

```python
emp['ename']
```

```python
emp.idx = emp['empno']
emp.idx
```

```python
emp.iloc[:,1:]
```

```python
emp = emp.set_index('ename') #'ename'을 인덱스로 지정
emp
```

```python
emp.sort_index(ascending = True) 
```

```python
emp.sort_index(axis = 0)  #행 기준으로 정렬
```

```python
emp.sort_index(axis = 1)   #column순 정렬
```

2. sort_values
- Series, DataFrame 호출 가능
- 본문의 값(value) 기준으로 정렬(Series, DataFrame 특정 컬럼 순서대로)

```python
emp.sort_values(by = 'sal')  #sal값에 의해 정렬  #by생략가능
```

```python
emp.sort_values(by = 'sal', ascending = False)
```

부서별 연봉 수준

```python
emp.sort_values(by = ['deptno','sal'])
```

연봉만 높은순으로 정렬

```python
emp.sort_values(by = ['deptno','sal'],ascending = [True, False])
```
