```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
```

# 선 그래프

```python
plt.plot([1,2,3,4],
         [10,20,30,40],
         marker = '^',
         linestyle = '--',
         color = 'r')
```

```python
s1 = Series([10,20,30,40])
s1.plot()  #산점도 생성
```

```python
s1.plot(xticks=[0,1,2,3],   #눈금좌표
        ylim=[0,100],       #y축 범위
        xlabel='x name',    #x축 라벨
        ylabel='y name',    #y축 라벨
        rot=90,             #rot (=rotation 회전) 90도
        title='name',       #title 그림 제목
        marker='^',         #마커  
        linestyle='--',     #선스타일
        color='red')        #컬러
```
